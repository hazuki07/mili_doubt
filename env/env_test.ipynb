```python
import env
import numpy as np
```

```python
env = env.MillionDoubtEnv()
# env.game.start_game()

obs = env.reset()
```

```python
log_episode_reward = []
log = []
obs = env.reset()

# MLPlayer vs human, CPUPlayer
while True:
    obs = env.update_obs()
    act = env.get_action(obs)
    next_obs, reward, done, info = env.step(act)
    
    log.apeend((obs, act, next_obs, reward, done))
    log_episode_reward.append(reward)
    
    if done:
        break
    
    obs = next_obs
    
episode_rewrd = sum(log_episode_reward)
```
